```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import numpy as np
import requests
from PIL import Image
import cv2
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
from io import BytesIO
```

# Gaussian Filter

```python
# Step 1: Get the image from the URL
img_url = "http://farm8.staticflickr.com/7218/7048835433_97d58c41dd_z.jpg"
response = requests.get(img_url)
img = Image.open(BytesIO(response.content))

# Step 2: Display the image
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
# gaussian blur
img = cv2.GaussianBlur(np.array(img), (5, 5), 0)
plt.imshow(img)

# save the image
cv2.imwrite('gaussian_filter.jpg', img)
```

```python
# Step 1: Get the image from the URL
img_url = "http://farm8.staticflickr.com/7218/7048835433_97d58c41dd_z.jpg"
response = requests.get(img_url)
img = Image.open(BytesIO(response.content))
original_img = img

# apply gaussian blur
img = cv2.GaussianBlur(np.array(img), (5, 5), 0)
img = Image.fromarray(img)

# originnal_img - gaussian_img = difference_img
# make gray scale original image
original_img = original_img.convert('L')
# make gray scale gaussian image
img = img.convert('L')
# original_img - gaussian_img = difference_img
difference_img = Image.fromarray(np.array(original_img) - np.array(img))


# display the images
fig, ax = plt.subplots(1, 3, figsize=(15, 15))
ax[0].imshow(original_img)
ax[0].set_title('Original Image')
ax[0].axis('off')
ax[1].imshow(img)
ax[1].set_title('Gaussian Image')
ax[1].axis('off')
ax[2].imshow(difference_img)
ax[2].set_title('Difference Image')
ax[2].axis('off')
plt.show()
```

# Canny Edge Detector

```python
# Step 1: Get the image from the URL
img_url = "http://farm8.staticflickr.com/7218/7048835433_97d58c41dd_z.jpg"
response = requests.get(img_url)
img = Image.open(BytesIO(response.content))

# Step 2: Display the image
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
# cann edge detection
img = cv2.Canny(np.array(img), 300, 700)
plt.imshow(img)

# save the image
cv2.imwrite('canny_edge.jpg', img)
```
