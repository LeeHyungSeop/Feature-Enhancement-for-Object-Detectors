# Figure 3 : baseline vs. cog

cog means S2 for 'Canny filtered image', S3 for 'Original image', S4 for 'Gaussian filtered image'

```python
import pickle
import matplotlib.pyplot as plt
```

## (a) loss, regression loss, classification loss graph

```python
baseline_loss = pickle.load(open('baseline_loss.pkl', 'rb'))
baseline_reg_loss = pickle.load(open('baseline_reg_loss.pkl', 'rb'))
baseline_cls_loss = pickle.load(open('baseline_cls_loss.pkl', 'rb'))

cog_loss = pickle.load(open('cog_loss.pkl', 'rb'))
cog_reg_loss = pickle.load(open('cog_reg_loss.pkl', 'rb'))
cog_cls_loss = pickle.load(open('cog_cls_loss.pkl', 'rb'))
```

```python
# Plotting total_loss, reg_loss, cls_loss in one figure
# baseline : dashed line and circle marker
# cog : solid line and circle marker

plt.figure()

plt.plot(baseline_reg_loss, 'r', label='baseline reg loss')
plt.plot(cog_reg_loss, 'r:', label='cog reg loss')

plt.plot(baseline_cls_loss, 'b', label='baseline cls loss')
plt.plot(cog_cls_loss, 'b:', label='cog cls loss')

plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.title('Comparison loss between baseline and cog')
plt.savefig('CoG_loss_comparison.png')
```

## (b) mAP Large, Medium, Small

```python
baseline_large_AP = pickle.load(open('baseline_large_AP.pkl', 'rb'))
baseline_medium_AP = pickle.load(open('baseline_medium_AP.pkl', 'rb'))
baseline_small_AP = pickle.load(open('baseline_small_AP.pkl', 'rb'))

cog_large_AP = pickle.load(open('cog_large_AP.pkl', 'rb'))
cog_medium_AP = pickle.load(open('cog_medium_AP.pkl', 'rb'))
cog_small_AP = pickle.load(open('cog_small_AP.pkl', 'rb'))
```

```python
# Plotting AP for each class
# baseline : dashed line and circle marker
# cog : solid line and circle marker

plt.figure()

plt.plot(baseline_large_AP, 'r', label='baseline large AP')
plt.plot(cog_large_AP, 'r:', label='cog large AP')

plt.plot(baseline_medium_AP, 'b', label='baseline medium AP')
plt.plot(cog_medium_AP, 'b:', label='cog medium AP')

plt.plot(baseline_small_AP, 'g', label='baseline small AP')
plt.plot(cog_small_AP, 'g:', label='cog small AP')

plt.xlabel('Epoch')
plt.ylabel('AP')
plt.legend()
plt.title('Comparison AP between baseline and cog')
plt.savefig('CoG_AP_comparison.png')
```
